### get_dummies method

```python
cds = ds.copy()
cds["Gender"].unique()
```

```python
dummies = pd.get_dummies(cds["Gender"], prefix="Gender_") # 
```

```python
dummies
```

```python
dummies = dummies.astype(int)
dummies
```

```python
cds = pd.concat([cds,dummies], axis=1)  
# axis=1 mean perform concatenation column wise
```

```python
cds
```

### one hot encoder

```python
ds["Married"].fillna(ds["Married"].mode()[0], inplace=True)
ds.isnull().sum()
```

```python
new_data = ds[["Gender","Married"]]
```

```python
new_data
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder() # creating object of oneHotEncoder
```

```python
ar = ohe.fit_transform(new_data).toarray()
ar
```

```python
pd.get_dummies(new_data).info()
```

```python
after_label_ar = pd.DataFrame(ar, columns=["Gender_Female", "Gender_Male", "Married_No","Married_Yes"])
after_label_ar
```
